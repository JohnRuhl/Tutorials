```python
import numpy as np
import yaml
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
plt.rcParams.update({'font.size': 18})
plt.rcParams['figure.figsize'] = [10, 6]
```

# Line of N discrete oscillators
The phase difference between each, at angle theta is:
$$ \delta = 2 \pi \sin\theta \frac{d}{\lambda} $$

```python
N = 10               # number of oscillators
d_over_lambda = 20  # spacing between oscillators, in wavelengths

theta_max = 45      #deg
theta_deg = np.linspace(-theta_max,theta_max, 10000)
theta_rad = theta_deg*np.pi/180.

delta = 2*np.pi*np.sin(theta_rad)*d_over_lambda

I = np.sin(N*delta/2)**2 / np.sin(delta/2)**2

plt.plot(theta_deg,I)
plt.grid()
```

# Continuous line of oscillators
Length of line is $D$, 

$$ \beta = \frac{\pi D}{\lambda} \sin\theta $$

$$ I(\theta) = I_0 \left(\frac{\sin\beta}{\beta}\right)^2 $$

```python
D_over_lambda = 6  # spacing between oscillators, in wavelengths
theta_max = 45      #deg
theta_deg = np.linspace(-theta_max,theta_max, 10000)
theta_rad = theta_deg*np.pi/180.

beta = np.pi*D_over_lambda*np.sin(theta_rad)

I = (np.sin(beta)/ beta)**2
plt.plot(theta_deg,I)
plt.grid()
```

# Double slit
Each slit has width $b$, slits are separated by distance $a$.
$$ \alpha = \frac{\pi a}{\lambda} \sin\theta $$
$$ \beta  = \frac{\pi b}{\lambda} \sin\theta $$
$$ I(\theta) = \left(\frac{\sin\beta}{\beta}\right)^2 \cos^2\alpha $$

```python
a_over_lambda = 20  # spacing between slits, in wavelengths
b_over_lambda = 5   # width of each slit, in wavelengths
theta_max = 45      #deg
theta_deg = np.linspace(-theta_max,theta_max, 10000)
theta_rad = theta_deg*np.pi/180.

alpha = np.pi*a_over_lambda*np.sin(theta_rad)
beta =  np.pi*b_over_lambda*np.sin(theta_rad)

I = (np.sin(beta)/ beta)**2 * np.cos(alpha)**2
plt.plot(theta_deg,I)
#plt.ylim(0,0.1)
plt.grid()
```

# Diffraction grating
Line of discrete oscillators, three wavelengths

```python
N = 20               # number of oscillators
d_over_lambda = 20  # spacing between oscillators, in wavelengths

theta_max = 1      #deg
theta_deg = np.linspace(-theta_max,theta_max, 10000)
theta_rad = theta_deg*np.pi/180.

colors = ['b','g','r']
ii = 0
for lamfac in [0.9,1.0,1.1]:
  delta = 2*np.pi*np.sin(theta_rad)*(d_over_lambda/lamfac)
  I = np.sin(N*delta/2)**2 / np.sin(delta/2)**2
  plt.plot(theta_deg,I,colors[ii])
  ii = ii+1

plt.xlabel('Degrees')
plt.grid()
```
