# Mechanical design calculations

Circular plate, stress and deflection.
Uniform load, edges simply supported.
Formulas from:  https://roymech.org/Useful_Tables/Mechanics/Plates.html

Center stress:
$$ \sigma_m = \frac{3(3+\nu)p r^2}{8 t^2}$$

Center deflection:
$$ y_m = \frac{(5+\nu)pr^4}{64 (1+\nu) D} $$

where:
$$ D = \frac{E t^3}{12 (1-\nu^2)} $$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size': 18})
plt.rcParams['figure.figsize'] = [10, 6]
```

Let's calculate for aluminum, r = 0.25m, over various t.

```python
# Aluminum
nu = 0.33
p = 101.e3 # N/m^2
E = 68e9 # Pa, N/m^2
r = 0.25 # m
t_inches = np.arange(0.25,0.5,0.01)
t_mm = t_inches*25.4
t = t_mm/1000.  # meters
D = (E*t**3)/(12*(1-nu**2))

# stress
sigma_m = 3*(3+nu)*p*r**2/(8*t**2)

# deflection
y_m = (5+nu)*p*r**4/(64*(1+nu)*D)

plt.rcParams['figure.figsize'] = [10, 8]
plt.subplot(2,1,1)
plt.title('Circular aluminum plate, 0.25m radius, resting at edge')
plt.plot(t_mm,1000*y_m)
plt.ylabel('Y deflection (mm)')
#plt.xlabel('Plate thickness (mm)')
plt.grid()

plt.subplot(2,1,2)
plt.plot(t_mm,sigma_m/1e6,label='Max stress')
plt.plot(t_mm,t_mm*0+240,'r',label='Al 6061-T6 yield stress')
plt.ylabel('Stress (MPa)')
plt.xlabel('Plate thickness (mm)')
plt.ylim(bottom=0)
plt.legend()
plt.grid()
```

# Aluminum plate, clamped at edge.
Circular plate, stress and deflection.
Uniform load, edges clamped.
Formulas from:  https://roymech.org/Useful_Tables/Mechanics/Plates.html

Center stress:
$$ \sigma_m = \frac{3 p r^2}{4 t^2}$$

Center deflection:
$$ y_m = \frac{pr^4}{64 D} $$

where:
$$ D = \frac{E t^3}{12 (1-\nu^2)} $$

```python
# Aluminum
nu = 0.33
p = 101e3 # N/m^2
E = 68e9 # Pa, N/m^2
r = 0.25 # m
t_inches = np.arange(0.25,0.5,0.01)
t_mm = t_inches*25.4
t = t_mm/1000.  # meters
D = (E*t**3)/(12*(1-nu**2))

# stress
sigma_m = 3*p*r**2/(4*t**2)

# deflection
y_m = p*r**4/(64*D)

plt.rcParams['figure.figsize'] = [10, 8]
plt.subplot(2,1,1)
plt.title('Circular aluminum plate, 0.25m radius, clamped at edge')
plt.plot(t_mm,1000*y_m)
plt.ylabel('Y deflection (mm)')
#plt.xlabel('Plate thickness (mm)')
plt.grid()

plt.subplot(2,1,2)
plt.plot(t_mm,sigma_m/1e6,label='Max stress')
plt.plot(t_mm,t_mm*0+240,'r',label='Al 6061-T6 yield stress')
plt.ylim(bottom=0)
plt.ylabel('Stress (MPa)')
plt.xlabel('Plate thickness (mm)')
plt.legend()
plt.grid()
```
