```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size': 12})
plt.rcParams['figure.figsize'] = [8, 5]
```

```python
theta = np.linspace(0,2*np.pi,1000)
y = 0
for ii in range(1000):
    nn = 2*ii+1
    y = y + (1/nn)*np.sin(nn*theta)
    plt.plot(theta,(4/np.pi)*y)
plt.grid()
plt.plot(theta,(4/np.pi)*y,'k',linewidth=3)
```

```python
thetas = np.linspace(0,8*np.pi,5000)
for theta in thetas:
    y = np.sin(theta)
    plt.scatter(theta,y,s=2,c='b')
```
