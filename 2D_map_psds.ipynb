# 2D Map fft's, psds, and power spectra.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size': 18})
plt.rcParams['figure.figsize'] = [10, 8]
```

```python
nside = 1024
psize =0.25
noiselevel = 10. # uK-arcmin
sigma_pix = noiselevel/psize

noise_map = np.random.normal(0,sigma_pix,[nside,nside])
print(np.std(noise_map))

plt.imshow(noise_map)
plt.colorbar()
```

```python
noise_fft = np.fft.fft2(noise_map,norm='ortho')
noise_psd = np.abs(noise_fft)**2
```
