```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.rcParams.update({'font.size': 18})
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
def drawcircle(x,y,r,n):
    theta = np.arange(0,2*np.pi,0.01)
    xvec = x + r*np.cos(theta)
    yvec = y + r*np.sin(theta)
    plt.text(x,y,n,horizontalalignment='center', verticalalignment='center')
    plt.plot(xvec,yvec,'b')
    
def drawcircleB(x,y,s,color):
    plt.scatter(x,y,s,c=color)
    
def drawhex(x,y,r):
    theta = np.array([0,60,120,180,240,300,360])
    theta = theta + 30
    theta = theta*np.pi/180
    dpoints = r/np.cos(30*np.pi/180)
    xvec = x + dpoints*np.cos(theta)
    yvec = y + dpoints*np.sin(theta)
    plt.plot(xvec,yvec,'r')

def drawsquare(x,y,r):
    theta = np.array([45,135,225,315,45])
    theta = theta*np.pi/180
    dpoints = r/np.cos(30*np.pi/180)
    xvec = x + dpoints*np.cos(theta)
    yvec = y + dpoints*np.sin(theta)
    plt.plot(xvec,yvec,'r')
```

```python
plt.clf()
nside = 6
diam = 2
nx = nside-1
x0 = diam*(-(nside-2)/2)
y0 = diam*(-(nside-1)*np.cos(30*np.pi/180))
n = 1
for row in range(2*nside-1):
    if row<=nside-1:
        nx = nx + 1
        x0 = x0 - 0.5*diam
    else:
        nx = nx - 1
        x0 = x0 + 0.5*diam
    for xx in range(nx):
        x = x0 + xx*diam
        y = y0 + row*np.cos(30*np.pi/180)*diam
        #drawcircle(x,y,0.5*diam,n)
        drawcircleB(x,y,500,'r')
        n=n+1
        #drawhex(x,y,0.5*diam)
        #drawsquare(x,y,diam/2)
xy =12
plt.xlim(-xy,xy)
plt.ylim(-xy,xy)
plt.grid()
```
